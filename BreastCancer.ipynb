### LOADING THE DATASET

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
print(df.head())
```

### DATA SPLITTING

```python
# import the dependancy
!pip install tensorflow
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = df.drop(columns=['target'])
y = df['target']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

### NEURAL NETWORK

```python
model = Sequential()
model.add(Dense(11,activation='relu',input_dim=30))   # Input
model.add(Dense(11,activation='relu'))             # hidden - 1
model.add(Dense(22,activation='relu'))             # hidden - 2
model.add(Dense(11,activation='relu'))             # hidden - 2
model.add(Dense(1,activation='sigmoid'))              # output
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100,validation_split = 0.2)
```

```python
train_pred = model.predict(x_train)
train_pred
```

```python
train_pred = np.where(train_pred>0.5,1,0)
print(train_pred)
```

```python
test_pred = model.predict(x_test)
test_pred = np.where(test_pred>0.5,1,0)
test_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print('Train acc :- ', accuracy_score(y_train,train_pred))
print('Test confusion_matrix :- \n', confusion_matrix(y_test,test_pred))
```

```python
print('Test acc :- ', accuracy_score(y_test,test_pred))
print('Test confusion_matrix :- \n', confusion_matrix(y_test,test_pred))
```

```python
prediction = model.predict(x_test)
prediction

if prediction[0] == 1:
    print("The tumor is **benign**.")
else:
    print("The tumor is **malignant**.")
```
